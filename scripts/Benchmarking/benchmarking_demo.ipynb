## Constant Source in Divergence form

The solution \( u(x, y) \) for the Poisson equation
\begin{equation}
\frac{\partial^2 u}{\partial x^2} + \frac{\partial^2 u}{\partial y^2} = 4
\end{equation}

with the boundary condition

\begin{equation}
u(x, y) = x^2 + y^2
\end{equation}

has analytical solution:

\begin{equation}
u(x, y) = x^2 + y^2
\end{equation}


```python
compare_model_to_benchmark("const_source", print_PDE=True)
```

## Constant Source in Non Divergence form

$$A(x,y):H(x,y) = 4$$

Where
\begin{equation}
A(x, y) = \begin{pmatrix}
1 & 0 \\
0 & 1 \\
\end{pmatrix}
\quad
\text{and}
\quad
H(x, y) = \begin{pmatrix}
\frac{\partial^2 u}{\partial x^2} & \frac{\partial^2 u}{\partial x \partial y} \\
\frac{\partial^2 u}{\partial y \partial x} & \frac{\partial^2 u}{\partial y^2}
\end{pmatrix}
\end{equation}

with the boundary condition

\begin{equation}
u(x, y) = x^2 + y^2
\end{equation}

has analytical solution:

\begin{equation}
u(x, y) = x^2 + y^2
\end{equation}



```python
compare_model_to_benchmark("const_source_NonDF", print_PDE=True)
```

## Eigenfunction source in Divergence form

I will try to solve the following poisson problem
\begin{equation}
\Delta u = f(x, y)
\end{equation}

with boundary conditions:

\begin{equation}
u(0, y) = u(1, y) = u(x, 0) = u(x, 1) = 0
\end{equation}

where

\begin{equation}
f(x, y) = -\sin\left(\pi x\right) \sin\left(\pi y\right)
\end{equation}

The analytical solution to this is
\begin{equation}
u(x, y) = \frac{1}{2 \pi^2} \sin\left(\pi x\right) \sin\left(\pi y\right)
\end{equation}

```python
compare_model_to_benchmark("eigenfunc", print_PDE=True)
```
